### **Exercise 4 â€” Comparing Modules**



1. Compute the **mean** and **standard deviation** of attendance per module using pandas `groupby`. Consult the documentation if needed.


````python
module_attendance_stats = df.groupby(['Module Name'], as_index=False).agg({'Attended':['mean','std']})
module_attendance_stats.columns = ['Module Name', 'Mean Attendance', 'Std Dev Attendance']
````

2. Create a `seaborn.barplot` of mean attendance per module with 95% confidence intervals.


````python
import seaborn as sns
````

````python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12,6))
sns.barplot( data=df,x="Module Name", y="Attended",estimator=np.mean,errorbar=('ci', 95))
plt.xticks(rotation=90)
plt.title("Average Attendance by Module with 95% Confidence Intervals")
plt.show()
````

### **Exercise 5 â€” Attendance Relationships**



1. Answer the question: **Do students with low overall attendance tend to attend modules that have low attendance?**.

2. **[For L5 students only]**. Choose between a correlation analysis or a linear regression analysis to answer the question.

3. **[For L6 students only]** Run both correlation analysis and linear regression analysis to answer the same question. Do they tell the same story?

> Tips:
> 1. Compute the mean attendance rate of each module. Can you re-use some of the `groupby` code you wrote above?
> 2. For each module, compute the average overall attendance of the students who attended that module.
> 3. Correlation and linear regression can both help you answer the question. 
> 4. Report main statistic and p-value(s).


````python

student_module = df[['Person Code', 'Module Name', 'Student Overall Attendance']].drop_duplicates(subset=['Person Code', 'Module Name'])
module_student_overall = student_module.groupby('Module Name')['Student Overall Attendance'].mean().reset_index(name='Mean Student Overall Attendance')
merged_df = module_attendance_stats.merge(module_student_overall, on='Module Name')
````

````python
from scipy.stats import pearsonr
r, p_value = pearsonr(merged_df['Mean Attendance'], merged_df['Mean Student Overall Attendance'])
print(r)
print(p_value)
````

````python
from scipy.stats import linregress
x = merged_df['Mean Attendance']
y = merged_df['Mean Student Overall Attendance']
lr = linregress(x, y)
print(lr)
````

````python
import plotly.express as px
sns.regplot(data=merged_df, x='Mean Attendance', y='Mean Student Overall Attendance')
fig = px.scatter(merged_df, x='Mean Attendance', y='Mean Student Overall Attendance', trendline='ols', title='Mean Module Attendance vs Mean Student Overall Attendance', hover_data=['Module Name'])
fig.update_traces(marker=dict(size=10))
fig.show()
````

````python
results = pd.DataFrame({
    'Metric': ['Pearson Correlation Coefficient', 'P-value', 'Regression Slope', 'Regression Intercept', 'R-squared'],
    'Value': [r, p_value, lr.slope, lr.intercept, lr.rvalue**2]
})
print(results)
````
